## 1. 시험장 분위기 및 시험 문제 복원+답변을 추출

```python
def extract_patterns(content):
    # 2번 시험장 분위기 추출
    exam_env_match = re.search(r'시험장 분위기.*?[:\n]\s*(.*?)(?:3\.\s*시험 문제 복원|$)', content, re.DOTALL)
    exam_env = exam_env_match.group(1).strip() if exam_env_match else ''

    # 3번 시험 문제 복원과 답변 추출
    qa_match = re.search(r'시험 문제 복원.*?[:\n]\s*(.*?)(?:\n\s*-\s*|$)', content, re.DOTALL)
    qa_content = qa_match.group(1).strip() if qa_match else ''

    return exam_env, qa_content

extracted_data = []

# Content 컬럼에서 시험장 분위기와 시험 문제 복원+답변 추출
for index, row in df.iterrows():
    content = row['Content']
    exam_env, qa_content = extract_patterns(content)

    # 추출한 데이터를 저장
    extracted_data.append({
        'Keyword': row['Keyword'],
        'Link': row['Link'],
        'Title': row['Title'],
        'Nickname': row['Nickname'],
        'Date': row['Date'],
        'Content': row['Content'],  # Content 컬럼 포함
        '시험장 분위기': exam_env,
        '시험 문제 복원 및 답변': qa_content
    })

extracted_df = pd.DataFrame(extracted_data)
extracted_df
```

```python
extracted_df.to_csv('extracted_NAVER_Cafe_crawling.csv', index=False)
```

```python
# 두 컬럼 중 하나라도 NaN이거나 빈 값인 데이터를 필터링
nan_df = extracted_df[
    extracted_df['시험장 분위기'].isna() |
    extracted_df['시험 문제 복원 및 답변'].isna() |
    (extracted_df['시험장 분위기'].str.strip() == '') |
    (extracted_df['시험 문제 복원 및 답변'].str.strip() == '')
]

# 둘 다 내용이 있는 데이터만 필터링
new_df = extracted_df.dropna(subset=['시험장 분위기', '시험 문제 복원 및 답변'])
new_df = new_df[(new_df['시험장 분위기'].str.strip() != '') &
                (new_df['시험 문제 복원 및 답변'].str.strip() != '')]


nan_df.reset_index(drop=True, inplace=True)


nan_df.to_csv('nan_NAVER_Cafe_crawling.csv', index=False)


new_df.to_csv('new_NAVER_Cafe_crawling.csv', index=False)
```

## 2. 질문과 답변을 추출

```python
def extract_qa_pairs(content):
    qa_list = []
    exam_env = ''

    # 시험장 분위기 추출: 2번에서 3번 항목이 나오기 전까지
    exam_env_match = re.search(r'2\.\s*시험장 분위기.*?\n(.*?)(?:3\.\s*시험 문제 복원|$)', content, re.DOTALL)
    if exam_env_match:
        exam_env = exam_env_match.group(1).strip()

    # 질문 패턴 정의
    question_patterns = [
        r".*?\?$",
        r"(?:어떤|어떻게|왜|무엇|무슨|언제|어디|누가|누구|몇|어느|요약|설명|이유).*?\?",
        r"(?:말씀해|생각해|알려|보여|들려|이해해|묘사해|논해|설명해|자기소개).*?\.",
        r"Q\d*\.*\s*[^\n]+",
        r"\b(?:질문|Q|Question|질문[0-9]|[0-9]+[.]|[0-9]+\))\b.*?"  # 다양한 질문 형태 포함
    ]

    current_question = None
    current_answer = []
    lines = content.split('\n')
    line_index = 0

    while line_index < len(lines):
        line = lines[line_index].strip()

        # 질문 패턴 감지
        is_question = False
        for pattern in question_patterns:
            if re.match(pattern, line):
                # 이전 질문-답변 쌍 저장
                if current_question and current_answer:
                    qa_list.append((exam_env, current_question, ' '.join(current_answer).strip()))
                # 새로운 질문 감지
                current_question = line
                current_answer = []
                is_question = True
                break

        # 질문이 아닌 경우, 현재 질문이 있다면 답변으로 간주
        if not is_question and current_question:
            current_answer.append(line)

        line_index += 1

    # 마지막 질문-답변 쌍 추가
    if current_question and current_answer:
        qa_list.append((exam_env, current_question, ' '.join(current_answer).strip()))

    return qa_list


# Content 항목에서 시험장 분위기와 질문-답변 추출
all_qa_pairs = []
for index, row in new_df.iterrows():
    content = row['Content']  # 'Content'에서 추출 시작
    qa_pairs = extract_qa_pairs(content)

    # 각 질문-답변 쌍에 기존 컬럼(닉네임, 링크 등)을 추가하여 확장
    for qa_pair in qa_pairs:
        extended_row = {
            'Keyword': row['Keyword'],
            'Link': row['Link'],
            'Title': row['Title'],
            'Nickname': row['Nickname'],
            'Date': row['Date'],
            '시험장 분위기': qa_pair[0],
            '질문': qa_pair[1],
            '답변': qa_pair[2]
        }
        all_qa_pairs.append(extended_row)


final_df = pd.DataFrame(all_qa_pairs)

final_df
```

```python
# '시험장 분위기', '질문', '답변' 컬럼 중 하나라도 NaN이거나 공백인 행 제거
filtered_df = final_df.dropna(subset=['질문', '답변'])  # NaN 값 제거
filtered_df = filtered_df[
    (filtered_df['질문'].str.strip() != '') &
    (filtered_df['답변'].str.strip() != '')
]

result_df = filtered_df.reset_index(drop=True)
result_df
```

## 3. 인터뷰 한묶음으로 전처리 코드

```python
result_data = []

# 현재 처리 중인 Nickname 및 history 초기화
current_nickname = None
history = ""

# 데이터프레임을 순회하면서 history-response pairs 구성
for index, row in result_df.iterrows():
    nickname = row['Nickname']
    question = row['질문']
    answer = row['답변']

    # 새로운 Nickname으로 넘어가면 history를 초기화
    if nickname != current_nickname:
        current_nickname = nickname
        history = ""  # 이전 인터뷰 세션의 history 초기화

    # 질문을 history에 추가
    if history:  # 기존 history가 있으면 공백을 추가하여 구분
        history += f" <U> {question}"
    else:
        history = f"<U> {question}"  # 첫 질문인 경우

    # 답변을 함께 저장하고 나서 history 갱신
    result_data.append({
        'Nickname': nickname,
        'history': history + " <S> ",
        'completion': answer
    })

    # 답변을 history에 추가하여 대화 기록을 갱신
    history += f" <S> {answer}"

result_df = pd.DataFrame(result_data, columns=['Nickname', 'history', 'completion'])
result_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
# prompt와 completion을 분리한 후, 'split' 컬럼 추가
result_df['prompt'] = result_df['history'].apply(lambda x: f"<|im_start|>user\n{x}<|im_end|><|im_start|>assistant\n")
result_df['completion'] = result_df['completion'].apply(lambda x: f"{x}<|im_end|>")

train_df, valid_df = train_test_split(result_df, test_size=0.2, random_state=42)

train_df['split'] = 'train'
valid_df['split'] = 'valid'

train_df = train_df[['prompt', 'completion', 'split']]
valid_df = valid_df[['prompt', 'completion', 'split']]

train_df.to_csv('Training_df_processed_sessions.csv', index=False, encoding='utf-8-sig')
valid_df.to_csv('Validation_df_processed_sessions.csv', index=False, encoding='utf-8-sig')

print(f"Training data size: {len(train_df)}")
print(f"Validation data size: {len(valid_df)}")
```

## 5. 모의면접 챗봇 Test

```python
conversation_history = []

max_conversation_pairs = 8

# 질문을 동적으로 생성하기 위한 템플릿 프롬프트
dynamic_prompt_template = """
당신은 컴퓨터학과 입시 면접관입니다.
지원자의 답변을 기반으로 한 번에 하나의 질문만 해주세요. 그리고 대화가 끊기지 않도록 유의해주세요.
현재까지의 대화 내용은 다음과 같습니다:
{conversation}

이제 다음 질문을 한 가지 해주세요.
"""

def generate_response(user_input, adapter_id):
    # 대화 기록을 이용하여 새로운 입력 프롬프트를 생성
    conversation = ""

    # 이전 대화 기록을 순차적으로 누적하여 새로운 프롬프트 생성
    for message in conversation_history:
        conversation += f"면접관: {message['assistant']}\n" if message['role'] == 'assistant' else f"지원자: {message['user']}\n"

    # 새로운 사용자 입력 추가 (이번 대화에서 유저가 입력한 내용)
    conversation += f"지원자: {user_input}\n"

    # 새로운 질문을 생성하기 위한 템플릿 기반 입력 프롬프트 생성
    input_prompt = dynamic_prompt_template.format(conversation=conversation)

    # 모델 응답 생성
    lorax_client = pb.deployments.client("solar-1-mini-chat-240612")

    try:
        response = lorax_client.generate(input_prompt, adapter_id=adapter_id, max_new_tokens=1000)
        assistant_reply = response.generated_text.strip()  # 여백 제거

        # 사용자 입력과 모델의 응답을 대화 기록에 저장
        conversation_history.append({'role': 'user', 'user': user_input})
        conversation_history.append({'role': 'assistant', 'assistant': assistant_reply})

        # 대화 기록의 길이를 체크하고, 10개 질문-답변 이후에는 종료 메시지 출력
        if len(conversation_history) // 2 >= max_conversation_pairs:
            return "모의 면접이 종료되었습니다. 감사합니다."

        return assistant_reply

    except Exception as e:
        print(f"Error during generation: {e}")
        return "오류가 발생했습니다."
```

```python
# 대화 시뮬레이션 예시
print("챗봇과 대화를 시작합니다. 'exit'을 입력하면 종료됩니다.")
while True:
    user_input = input("You: ")

    if user_input.lower() == 'exit':
        break

    # 대화가 10개 쌍을 초과하면 종료
    if len(conversation_history) // 2 >= max_conversation_pairs:
        # print("Assistant: 모의 면접이 종료되었습니다. 감사합니다.")
        break
    response = generate_response(user_input, adapter_id)
    print(f"Assistant: {response}")
```
